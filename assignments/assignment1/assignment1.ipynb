## Task 1: A List of Dictionaries

```python
transactions = [
    {"id": 1, "product": "Laptop",   "price": 1200, "quantity":  2, "region": "EU"},
    {"id": 2, "product": "Phone",    "price":  800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet",   "price":    0, "quantity":  5, "region": "EU"},
    {"id": 4, "product": "Monitor",  "price":  350, "quantity":  3, "region": "US"},
    {"id": 5, "product": "Laptop",   "price": 1200, "quantity":  1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price":   75, "quantity": 10, "region": "APAC"},
]

# a) Filter out transactions with zero price or negative quantity
valid_transactions = [t for t in transactions if t["price"] > 0 and t["quantity"] > 0]

# b) Compute total revenue per transaction
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]

# c) Group by region and sum revenue
revenue_by_region = {}
for t in valid_transactions:
    region = t["region"]
    revenue_by_region[region] = revenue_by_region.get(region, 0) + t["total"]

# d) Print the region with the highest total revenue
top_region = max(revenue_by_region, key=revenue_by_region.get)
print(f"Top region: {top_region} with total revenue {revenue_by_region[top_region]}")
```

## Task 2: Creating a Function

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(sales):
    """Return summary statistics for a list of daily sales values."""
    total = sum(sales)
    average = round(total / len(sales), 2)
    best_day = sales.index(max(sales)) + 1
    worst_day = sales.index(min(sales)) + 1
    above_average_count = sum(1 for s in sales if s > average)

    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count,
    }

print(summarise_sales(sales))
```

## Task 3: Loops

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia",  "grade": 91, "attendance": 0.98},
    {"name": "Leo",  "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom",  "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# a) Names of students with grade >= 50
passing = [s["name"] for s in students if s["grade"] >= 50]
print("Passing students:", passing)

# b) Status message per student
for s in students:
    if s["grade"] >= 80:
        status = "Excellent"
    elif s["grade"] >= 50:
        status = "Pass"
    else:
        status = "Fail"
    print(f"{s['name']}: {status}")

# c) Count students with attendance below 0.80
low_attendance_count = sum(1 for s in students if s["attendance"] < 0.80)
print(f"\nStudents with attendance below 80%: {low_attendance_count}")

# d) Find student with highest grade using a loop
top_student = students[0]
for s in students[1:]:
    if s["grade"] > top_student["grade"]:
        top_student = s
print(f"\nTop student (loop): {top_student['name']} with grade {top_student['grade']}")

# e) Find student with highest grade using max()
top_student_max = max(students, key=lambda s: s["grade"])
print(f"Top student (max): {top_student_max['name']} with grade {top_student_max['grade']}")
```

## Task 4: Class and Methods

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    def add_asset(self, ticker, shares, price_per_share):
        """Add a new asset or update an existing one."""
        self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}

    def total_value(self):
        """Return the total market value of the portfolio."""
        return sum(
            asset["shares"] * asset["price_per_share"]
            for asset in self.holdings.values()
        )

    def most_valuable(self):
        """Return the ticker of the asset with the highest total value."""
        return max(self.holdings, key=lambda t: self.holdings[t]["shares"] * self.holdings[t]["price_per_share"])

    def __str__(self):
        return (
            f"Portfolio owner: {self.owner}\n"
            f"Number of assets: {len(self.holdings)}\n"
            f"Total value: €{self.total_value():,.2f}"
        )


# Demonstration
portfolio = Portfolio("Nils Osterloff")
portfolio.add_asset("AAPL", shares=10, price_per_share=185.0)
portfolio.add_asset("MSFT", shares=5,  price_per_share=420.0)
portfolio.add_asset("NVDA", shares=8,  price_per_share=875.0)

print(portfolio)
print(f"Most valuable asset: {portfolio.most_valuable()}")
print(f"Total portfolio value: €{portfolio.total_value():,.2f}")
```

## Task 5: Importing a Module

```python
from datetime import date

orders = [
    ("2025-01-15", "Laptop"),  ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"),  ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"),   ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"),  ("2025-04-05", "Monitor"),
]

# Parse date strings into date objects
parsed_dates = [date.fromisoformat(order[0]) for order in orders]

# Compute the number of days between earliest and latest order
earliest = min(parsed_dates)
latest = max(parsed_dates)
days_between = (latest - earliest).days

print(f"Earliest order: {earliest}")
print(f"Latest order:   {latest}")
print(f"Days between:   {days_between}")
```

## Task 6: Array Operations and Broadcasting

```python
import numpy as np

# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]
])

# a) Total annual revenue per year (row sums)
annual_revenue = revenue.sum(axis=1)
print("Annual revenue (2022–2024):", annual_revenue)

# b) Mean quarterly revenue across all years (column means)
quarterly_mean = revenue.mean(axis=0)
print("Mean quarterly revenue (Q1–Q4):", quarterly_mean)

# c) Apply growth adjustment to 2024 row via broadcasting
growth_factors = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth_factors
print("Adjusted 2024 revenue:", adjusted_2024)

# d) All quarterly values exceeding 250
above_250 = revenue[revenue > 250]
print("Quarterly revenues above 250:", above_250)
```

## Task 7: Pandas

```python
import pandas as pd

data = {
    "employee":          ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department":        ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary":            [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp":         [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}
df = pd.DataFrame(data)

# a) Add salary_per_year_exp column
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round().astype(int)

# b) Top 3 employees by performance score
top3 = df.sort_values("performance_score", ascending=False).head(3)
print("Top 3 by performance score:")
print(top3)

# c) Finance employees with performance score > 4.0
finance_high = df.loc[(df["department"] == "Finance") & (df["performance_score"] > 4.0)]
print("\nFinance employees with score > 4.0:")
print(finance_high)

# d) Rename years_exp to experience_years
df = df.rename(columns={"years_exp": "experience_years"})

# e) Salary and performance score for first four employees using iloc
# salary is column index 2, performance_score is column index 4
print("\nSalary and performance score for first 4 employees:")
print(df.iloc[:4, [2, 4]])
```

## Task 8: More Pandas

```python
import numpy as np
import pandas as pd

# Generate the data
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) Create DataFrame with date as index
df = pd.DataFrame({
    "date":            days,
    "visitors":        visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value,
}).set_index("date")

# b) Compute conversions and daily_revenue
df["conversions"]    = (df["visitors"] * df["conversion_rate"]).round().astype(int)
df["daily_revenue"]  = (df["conversions"] * df["avg_order_value"]).round(2)

print(df)
```
